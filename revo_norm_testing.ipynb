## Quick Test Function

This helper function makes it easy to test multiple sentences at once.

````python
def test_normalize(text, language="ms", config=None, extract_entities=False):
    """Test normalization and display results."""
    result = normalize_text(
        text,
        language=language,
        config=config,
        extract_entities_first=extract_entities
    )
    
    print(f"Input ({language.upper()}):  {text}")
    print(f"Output:              {result}")
    print()
    return result

def test_batch(test_cases, language="ms"):
    """Test multiple cases and display as a table."""
    results = []
    for input_text in test_cases:
        output = normalize_text(input_text, language=language)
        results.append({"Input": input_text, "Output": output})
    
    df = pd.DataFrame(results)
    display(HTML(df.to_html(index=False)))
    return results

def compare_configs(text, language="ms"):
    """Compare different configuration profiles."""
    configs = {
        "minimal": minimal_config(),
        "basic": basic_config(),
        "standard": standard_config(),
        "aggressive": aggressive_config()
    }
    
    print(f"Input: {text}\n")
    for name, config in configs.items():
        result = normalize_text(text, language=language, config=config)
        print(f"[{name.upper()}] {result}")
    print()

print("Test functions loaded!")
````

# 20. CUSTOM CONFIGURATION

Create custom configurations by enabling/disabling specific features.

````python
# Create custom config - disable acronym expansion
custom_config = standard_config()
custom_config.with_feature(FeatureGroup.ACRONYMS, FeatureLevel.OFF)

text = "Doktor dari KKM akan ke KTM"
print(f"Input: {text}\n")
print("[Standard - acronyms expanded]:")
print(normalize_text(text, language="ms", config=standard_config()))
print("\n[Custom - acronyms preserved]:")
print(normalize_text(text, language="ms", config=custom_config))

# Create minimal config - only basic cleanup
minimal = minimal_config()
print("\n[Minimal - basic cleanup only]:")
print(normalize_text(text, language="ms", config=minimal))
````
